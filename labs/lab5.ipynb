## Вариант 1

Реализуйте класс, описывающий космическую ракету. Назовите его ```Rocket```.

Из всех свойств реальной ракеты нас будут интересовать следующие:
* общая масса (кг)
* текущее количество топлива (1 литр = 1 кг)
* запущен ли двигатель (boolean)

Для создания ракеты нужно задать два числа - общую массу ракеты и вес топлива (входит в общую массу). Если топливо есть, то сразу должен включиться двигатель.

В процессе работы двигателя расходуется топливо, что влияет на его количество и общую массу ракеты. Как только топливо заканчивается, двигатель глохнет.
Время у объекта не реальное и управляется нами. Так что ракета не расходует топливо, пока мы это не скажем.

Для расхода топлива нужно сделать метод ```spend_fuel(count)```, который расходует ```count``` кг топлива и возвращает ```True```, если после этого топливо ещё осталось.
Если же топлива было недостаточно, то уровень топлива устанавливается в ```0```, а двигатель глохнет. При этом возвращается ```False```.

Текущее количества топлива можно узнать методом ```get_fuel_level()```, текущую общую массу - методом ```get_total_weight()```. Работает ли двигатель - методом ```get_is_engine_running()```.

При разработке класса соблюдите принцип инкапсуляции. public-члены предварите комментарием с документацией (напишите, что делает метод, что в него передавать, что он возвращает).

Сделайте ```main()```, который создаёт тестовую ракету и расходует топливо небольшими порциями, пока оно не кончится. Каждое состояние ракеты надо вывести на экран отдельной строкой, в которой указать общую массу, количество топлива, признак работы двигателя.
